```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from handle_dataset.transform import create_df_with_datetimes
```

```python
data_df = pd.read_csv('Dataset/Yearly-train.csv')
info_df = pd.read_csv('Dataset/M4-info.csv')
```

```python
first_timeseries = create_df_with_datetimes(data_df, 0)
first_timeseries['timestamp'] = pd.to_datetime(first_timeseries['timestamp'])
first_timeseries['year'] = first_timeseries['timestamp'].dt.year
plt.figure(figsize=(10,6))
plt.plot(first_timeseries['year'].tolist(), first_timeseries['target'].tolist())
plt.title('First timeseries')
plt.xlabel('Timestamp')
plt.ylabel('Target')
```

```python
random_timeseries = create_df_with_datetimes(data_df, 100)
random_timeseries['timestamp'] = pd.to_datetime(random_timeseries['timestamp'])
random_timeseries['year'] = random_timeseries['timestamp'].dt.year
plt.figure(figsize=(10,6))
plt.plot(random_timeseries['year'].tolist(), random_timeseries['target'].tolist())
plt.title('First timeseries')
plt.xlabel('Timestamp')
plt.ylabel('Target')
```
